As características de defasagem (lag features) são valores de períodos anteriores que são considerados úteis porque se baseiam na suposição de que o que aconteceu no passado pode influenciar ou conter informações intrínsecas sobre o futuro. #### Rolling Window Statistics

O objetivo principal de construir e usar estatísticas de janela móvel (rolling window statistics) em um conjunto de dados de série temporal é calcular estatísticas com base em valores de uma amostra de dados, definindo um intervalo que inclui a amostra em si, além de um número especificado de amostras antes e depois dela.

Um passo crucial ao calcular estatísticas móveis é definir uma janela móvel de observações: em cada ponto no tempo, os cientistas de dados obtêm as observações dentro da janela móvel e as usam para calcular a estatística desejada. Em seguida, movem-se para o próximo ponto no tempo e repetem a mesma computação na próxima janela de observações.

```python
# criar rolling mean feature

demand_val = demand_df[['daily_demand']]
shifted = demand_val.shift(1)
window = shifted.rolling(window=6)
means = window.mean()
new_dataframe = pd.concat([means, demand_val], axis=1)
new_dataframe.columns = ['daily_demand_rol_mean', 'daily_demand']
print(new_dataframe.head(10))
```

#### Expanding Window Statistics

As janelas expansivas (Expanding Window) são características que incluem todos os dados anteriores. A biblioteca pandas oferece a função `expanding()`, que realiza transformações expansivas e reúne todos os valores anteriores para cada instante de tempo. Python oferece a mesma interface e capacidades para as funções `rolling()` e `expanding()`.

```python
# criar expanding window features

demand_val = demand_df[['daily_demand']]
window = demand_val.expanding()
new_dataframe = pd.concat([window.min(),
window.mean(), window.max(), demand_val. shift(-1)], axis=1)
new_dataframe.columns = ['min', 'mean', 'max', 'daily_demand+1']
print(new_dataframe.head(10))
```

## **Cap. 4 - ARIMA**

### **Autoregressão (AR)**

A autoregressão é uma abordagem de previsão de séries temporais que depende apenas das saídas anteriores de uma série temporal: essa técnica assume que as observações futuras no próximo carimbo de tempo estão relacionadas às observações nos carimbos de tempo anteriores por meio de uma relação linear. Em outras palavras, em uma autoregressão, um valor de uma série temporal é regredido em valores anteriores dessa mesma série temporal. Neste capítulo do livro, aprenderemos como implementar uma autoregressão para previsão de séries temporais com Python. Na autoregressão, o valor de saída no carimbo de tempo anterior se torna o valor de entrada para prever o valor do próximo carimbo de tempo, e os erros seguem as suposições usuais sobre erros em um modelo de regressão linear simples.

Os cientistas de dados se referem a autoregressões nas quais o valor previsto atual (saída) é baseado no valor imediatamente anterior (entrada) como autoregressão de primeira ordem. Em autoregressões, o número de valores de entrada precedentes na série temporal que são usados para prever o valor do próximo carimbo de tempo é chamado de ordem (frequentemente nos referimos à ordem com a letra p). Esse valor de ordem determina quantos pontos de dados anteriores serão usados: geralmente, os cientistas de dados estimam o valor de p testando diferentes valores e observando qual modelo resultou no critério de informação de Akaike (AIC) mínimo. Vamos discutir tanto o critério de informação de Akaike quanto o critério de informação Bayesiano (BIC) penalizado mais tarde.

De maneira mais geral, uma autoregressão de ordem *n* é uma regressão linear múltipla na qual o valor da série em qualquer momento *t* é uma função linear dos valores anteriores na mesma série temporal. Devido a essa dependência serial, outro aspecto importante das autoregressões é a autocorrelação: a autocorrelação é uma propriedade estatística que ocorre quando uma série temporal está linearmente relacionada a uma versão anterior ou defasada de si mesma. A autocorrelação é um conceito crucial para autoregressões, pois quanto maior a correlação entre a saída (ou seja, a variável alvo que precisamos prever) e uma variável defasada específica (ou seja, um grupo de valores em um carimbo de tempo anterior usado como entrada), mais peso a autoregressão pode atribuir a essa variável específica, considerando-a como tendo um forte poder preditivo.

Além disso, alguns métodos de regressão, como regressão linear e regressão de mínimos quadrados ordinários, dependem da suposição implícita de que não há presença de autocorrelação no conjunto de dados de treinamento usado para alimentar o modelo. Esses métodos, como regressão linear e regressão de mínimos quadrados ordinários, são definidos como metodologias paramétricas, pois o conjunto de dados usado com eles apresenta uma distribuição normal e sua função de regressão é definida em termos de um número finito de parâmetros desconhecidos que são estimados a partir dos dados. Por todas essas razões, a autocorrelação pode ajudar os cientistas de dados a selecionar o método mais apropriado para suas soluções de previsão de séries temporais. Além disso, a autocorrelação pode ser muito útil para obter insights adicionais de seus dados e entre suas variáveis e identificar padrões ocultos, como sazonalidade e tendência em dados de séries temporais. Para verificar se há presença de autocorrelação em seus dados de séries temporais, você pode usar dois gráficos integrados diferentes fornecidos pelo pandas, chamados `lag_plot` e `autocorrelation_plot`. Essas funções podem ser importadas de `pandas.plotting` e recebem uma Série ou DataFrame como argumento. Ambos os gráficos são verificações visuais que você pode aproveitar para ver se há autocorrelação em seu conjunto de dados de séries temporais.

O gráfico de autocorrelação mostra o valor da função de autocorrelação no eixo vertical, variando de -1 a 1. As linhas horizontais correspondem a faixas de confiança de 95% e 99%, com uma linha tracejada indicando a faixa de 99%. Esse gráfico é utilizado para determinar se os pontos de dados de uma série temporal são positivamente correlacionados, negativamente correlacionados ou independentes entre si. O gráfico da autocorrelação de uma série temporal em relação ao seu atraso é chamado de função de autocorrelação (ACF). O Python suporta a ACF com a função `plot_acf()` da biblioteca statsmodels. Abaixo está um exemplo de cálculo e plotagem do gráfico de autocorrelação para o conjunto de dados `ts_data_load` usando a função `plot_acf()` da biblioteca statsmodels.


Os intervalos de confiança são representados como um cone. Por padrão, isso é definido como um intervalo de confiança de 95%, sugerindo que valores de correlação fora desse cone são muito provavelmente uma correlação.

Outro conceito importante é a função de correlação parcial (PACF), que é uma correlação condicional. É a correlação entre duas variáveis sob a suposição de que consideramos os valores de um outro conjunto de variáveis. Em regressão, essa correlação parcial pode ser encontrada correlacionando os resíduos de duas regressões diferentes.

Em um conjunto de dados de séries temporais, a autocorrelação entre um valor em um instante de tempo e outro valor em um instante anterior consiste tanto na correlação direta entre esses dois valores quanto nas correlações indiretas. Essas correlações indiretas são uma função linear da correlação do valor em observação com valores em instantes intermediários.

O Python suporta a função PACF com o `plot_pacf()` da biblioteca statsmodels (statsmodels.org/stable/generated/statsmodels.graphics.tsaplots.plot_pacf.html). As funções de autocorrelação (ACF) e autocorrelação parcial (PACF) são essenciais na construção de modelos de autorregressão, que assumem que os valores atuais de uma série temporal podem ser obtidos a partir dos valores passados. O PACF ajuda a evitar multicolinearidade, identificando as defasagens mais relevantes. Por exemplo, se a ACF mostra boa correlação até a oitava defasagem, não é viável usar todas elas devido à multicolinearidade; por isso, usamos a PACF para selecionar as defasagens mais importantes.

No gráfico PACF, defasagens até 3 têm boa correlação antes de ultrapassar o intervalo de confiança, determinando o valor de p para o modelo autorregressivo. Embora incluir mais defasagens possa melhorar o ajuste do modelo, há risco de overfitting.


### **Moving Average (Média Móvel)**

A técnica de média móvel usa erros de previsão anteriores para prever futuras observações, sendo cada observação futura uma média ponderada desses erros. Junto com a autorregressão, forma a base dos modelos ARMA e ARIMA, que são mais sofisticados. Enquanto a autorregressão combina observações passadas, a média móvel combina erros passados, ajudando a explicar padrões ocultos nos dados. O Python oferece suporte a essa técnica através da classe ARMA no pacote statsmodels.

### **Autoregressive Moving Average (ARMA)**

Os modelos de média móvel autorregressiva (ARMA) são fundamentais na modelagem de séries temporais devido à sua estrutura linear, que simplifica a previsão linear (Zhang et al. 2015). Um método ARMA consiste em duas partes:

- Uma autorregressão
- Um modelo de média móvel

Comparados aos modelos autorregressivos e de média móvel, os modelos ARMA fornecem o modelo linear mais eficiente para séries temporais estacionárias, modelando o processo desconhecido com o mínimo de parâmetros (Zhang et al. 2015). Especificamente, os modelos ARMA são usados para descrever séries temporais estocásticas estacionárias semanais usando dois polinômios: um para a autorregressão e outro para a média móvel. Este método é frequentemente referido como modelo ARMA(p,q), onde:

- **p** é a ordem do polinômio autorregressivo
- **q** é a ordem do polinômio da média móvel

O Python suporta a implementação do modelo ARMA com a função `ARMA()` da biblioteca `statsmodels`. Esta biblioteca oferece a definição e parâmetros para a classe `ARMA()`.


### **Autoregressive Integrated Moving Average (ARIMA)**

Os modelos ARIMA são uma evolução dos modelos ARMA, incorporando a integração. Ambos utilizam autorregressão (AR) e média móvel (MA). A principal diferença é que ARIMA inclui diferenciação para tornar séries temporais estacionárias.

Os modelos ARIMA têm três parâmetros:
- **p**: ordem da autorregressão.
- **d**: grau de diferenciação.
- **q**: ordem da média móvel.

Para séries temporais sazonais, usa-se o modelo SARIMAX, que adiciona parâmetros para sazonalidade e fatores exógenos.

Python suporta ARIMA e SARIMAX através da biblioteca `statsmodels` [SARIMAX](https://www.statsmodels.org/dev/generated/statsmodels.tsa.statespace.sarimax.SARIMAX.html).

O "walk forward validation" é um método de validação para modelos de séries temporais. Ele envolve re-treinar o modelo a cada nova observação de dados, permitindo previsões mais precisas em cada passo temporal. Começando do início da série, o modelo é treinado com um conjunto de dados, faz uma previsão para o próximo passo e é avaliado. Isso é repetido para cada nova observação, proporcionando uma estimativa robusta do desempenho do modelo na prática. No entanto, isso requer o treinamento de muitos modelos, o que pode ser caro computacionalmente. Apesar disso, é considerado o padrão ouro para avaliação de modelos de séries temporais e é recomendado para projetos próprios.

```python
# Create a test data point for each HORIZON step
test_shifted = test.copy()
for t in range(1, HORIZON):
  test_shifted['daily_demand+'+str(t)] = test_shifted['daily_demand'].shift(-t, freq='D')

test_shifted = test_shifted.dropna(how='any')
```

Com base nos resultados obtidos, recomenda-se a continuidade da pesquisa com foco em aprimoramentos nos modelos de RNNs, bem como a investigação de outras técnicas de aprendizado profundo para potencializar ainda mais a precisão das previsões.

Este trabalho contribui para o campo da previsão de demanda energética ao demonstrar a aplicabilidade de métodos avançados de séries temporais e aprendizado de máquina. As recomendações práticas fornecidas podem auxiliar na implementação de sistemas de previsão mais eficientes, promovendo uma gestão mais eficaz dos recursos energéticos no Brasil.
